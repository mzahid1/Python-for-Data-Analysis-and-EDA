# A smart skill to master in Pandas - Pivot/Unpivot 

```python
notgtdataonly.columns
```

```python
notgtdataonly.to_csv('NOTGT.csv')
```

```python
import numpy as np

pivot_table2=pd.pivot_table(notgtdataonly, values='SalesPKR', aggfunc=np.sum,
                            index=['Customer Code','Customer Name'],
                            columns=['Sales Department'],
                            fill_value=0)
```

```python
display(pd.melt(pivot_table2.reset_index(),
                id_vars=['Customer Code','Customer Name'],
                var_name=['Division'],
                value_name='Sales'))
```
