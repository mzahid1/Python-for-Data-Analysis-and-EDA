```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#Load botgh weather data 
```

```python
df1=pd.read_csv('weatherKHI.csv',index_col=0)
df2=pd.read_csv('weatherLHR.csv',index_col=0)
df2.head()
```

```python
df1['City']='KHI'
df2['City']='LHR'
```

```python
df=pd.concat([df1,df2], axis=0)
```

```python
df['Date']=df.index
```

```python
df.info()
```

```python
print(df.columns)
```

```python
type(df)
```

```python
plt.plot(df['Sunshine'],marker='^',linestyle='--',color='r')
plt.plot(df['Evaporation'],color='g',linestyle='-')
plt.xlabel('Date')
plt.ylabel('Sunshine or Evaporation Amount')
plt.title('Sunshine Comparison')
plt.show()
```

```python
plt.plot(df['Sunshine'],df['Evaporation'])
plt.show()
```

```python
plt.scatter(df['Sunshine'],df['Evaporation'])
plt.ylabel('Evaporation',color='b', fontsize=25)
plt.xlabel('Sunshine',labelpad=5, fontsize=25, fontfamily='serif',color='r')
plt.title('Evaporation vs Sunshine', fontsize=25)
```
